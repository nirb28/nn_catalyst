```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("final_results.csv")  # Load your dataset

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['R2 Train', 'R2 Val', 'R2 Test', 'R2 All']])
plt.title("Distribution of R² Scores Across Splits")
plt.ylabel("R² Score")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df_sorted = df.sort_values('R2 Test', ascending=False)
sns.barplot(x='R2 Test', y='Name', data=df_sorted, palette='viridis')
plt.title("Models Sorted by Test R² Performance")
plt.xlabel("R² Score")
plt.ylabel("Target Name")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='R2 Train', y='R2 Test', hue='R2 All', size='R2 Val', sizes=(20, 200))
plt.plot([0, 1], [0, 1], 'k--', alpha=0.3)  # Diagonal line
plt.title("Train vs. Test Performance")
plt.show()
```

```python
plt.figure(figsize=(10, 12))
df_heat = df.set_index('Name')[['R2 Train', 'R2 Val', 'R2 Test']]
sns.heatmap(df_heat, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("R² Scores Across Targets and Splits")
plt.show()
```
